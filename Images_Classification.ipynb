```
# Standard scientific Python imports
import matplotlib.pyplot as plt
# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import PrecisionRecallCurve
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```
#import the dataset
digits = datasets.load_digits()
print(digits.data)
```

```
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('Training: %i' % label)
```

```
X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state=1)
```

##Precision-Recall Curves

####SVC

```
viz = PrecisionRecallCurve(SVC())
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

####Logistic Regression

```
viz = PrecisionRecallCurve(LogisticRegression())
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

####Ridge Classifier

```
# Create the visualizer, fit, score, and poof it
viz = PrecisionRecallCurve(RidgeClassifier())
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

####Random Forest

```
viz = PrecisionRecallCurve(RandomForestClassifier())
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

####K-Neighbors Classifier

```
viz = PrecisionRecallCurve(KNeighborsClassifier())
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.poof()
```

##Class Prediction Error

```
from yellowbrick.classifier import ClassPredictionError
```

####SVC

```
# Instantiate the classification model and visualizer
visualizer = ClassPredictionError(
    SVC()
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
g = visualizer.poof()
```

####Logistic Regression

```
# Instantiate the classification model and visualizer
visualizer = ClassPredictionError(
    LogisticRegression()
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
g = visualizer.poof()
```

####Ridge Classifier

```
# Instantiate the classification model and visualizer
visualizer = ClassPredictionError(
    RidgeClassifier()
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
g = visualizer.poof()
```

####Random Forest Classifier

```
# Instantiate the classification model and visualizer
visualizer = ClassPredictionError(
    RandomForestClassifier()
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
g = visualizer.poof()
```

####K-nearest Neighbors Classifier

```
# Instantiate the classification model and visualizer
visualizer = ClassPredictionError(
    KNeighborsClassifier()
)

# Fit the training data to the visualizer
visualizer.fit(X_train, y_train)

# Evaluate the model on the test data
visualizer.score(X_test, y_test)

# Draw visualization
g = visualizer.poof()
```

##**The Conclusion: the K-nearest Neighbors Classifier is the most precise model to deal with digits recognition, and the SVC is the worst one. Hence, when we want to deal with the digits classification problem, we can try k-nearest neighbors firstly**
